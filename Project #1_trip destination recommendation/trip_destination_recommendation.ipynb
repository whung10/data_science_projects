```python
# take San Francisco (SFO) for example and read all the price history; define a function to read flight data

filenames = ['~/Desktop/data_science_springboard/Project #1_trip destination recommendation/data/SFO/ATL_SFO.json',
             '~/Desktop/data_science_springboard/Project #1_trip destination recommendation/data/SFO/BOS_SFO.json',
             '~/Desktop/data_science_springboard/Project #1_trip destination recommendation/data/SFO/BWI_SFO.json',
             '~/Desktop/data_science_springboard/Project #1_trip destination recommendation/data/SFO/CLT_SFO.json',
             '~/Desktop/data_science_springboard/Project #1_trip destination recommendation/data/SFO/DCA_SFO.json']
#flight_dataset = {'flight_labels': ['monthly_flight_avg', 'monthly_flight_min', 'month'], 'flight_data': []}

def flight(filenames):
    """Get monthly average flight ticket price of each city and minimum price"""
    for f in filenames:
        print(f[-12:-5])
        
        df = pd.read_json(f)
        df_avg = df.groupby(['year']).mean()        
        print(df_avg)
        
        df_min = df_avg[df_avg['price'] == df_avg['price'].min()]
        print(df_min)
```

```python
flight(filenames)
```
